```python
import pandas as pd
import numpy as np
import os
```

```python
img_csv_path = './archive/indiana_projections.csv'
report_csv_path = './archive/indiana_reports.csv'
```

```python
img_df = pd.read_csv(img_csv_path, index_col='uid')
report_df = pd.read_csv(report_csv_path, index_col='uid')
```

```python
img_df.head()
```

```python
report_df = report_df.rename(columns = { 'impression': 'reports' })
report_df.head()
```

```python
# match the uid of img_df and report_df and add reports to img_df
img_df = img_df.join(report_df, on='uid')
img_df.head()
```

```python
img_df.shape
```

```python
img_df = img_df.dropna(subset=['reports'])
```
